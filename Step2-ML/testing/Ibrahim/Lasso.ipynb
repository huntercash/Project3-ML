```python
from sklearn.linear_model import LinearRegression, Lasso
import numpy as np
import pandas as pd
```

```python
x1 = np.array([1,4,6,4])
x2 = np.array([2,1,2,3])
x3 = np.array([3,2,6,2])
y = np.array([8,11,18,17])
```

```python
X = np.stack([x1,x2,x3], axis=1)
X
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_
```

```python
lasso = Lasso(alpha=1).fit(X,y)
```

```python
lasso.coef_
```

```python
lasso = Lasso(alpha=0.1).fit(X,y)
```

```python
lasso.coef_
```

```python
lasso = Lasso(alpha=0.01).fit(X,y)
lasso.coef_
```

```python
data = pd.read_csv('Cleaned 2013_2014.csv')
data
```

```python
Y = data['Nat Log Mean earnings of students working and not enrolled 10 years after entry']
Y.head()
# Median earnings of students working and not enrolled 10 years after entry
```

```python
X = data.drop(columns=['Nat Log Mean earnings of students working and not enrolled 10 years after entry','Unnamed: 14','Unnamed: 15'])
X.head()
```

```python
lasso = Lasso(alpha=1).fit(X,Y)
lasso
```

```python
lasso.coef_
```

```python
X.columns
```

```python
lasso_results = pd.DataFrame()
```

```python
lasso_results['feature']=X.columns
```

```python
lasso_results['coef']=lasso.coef_
```

```python
lasso_results
```
