```python
import pandas as pd
import os
```

```python
df1 = pd.read_csv(os.path.join('machine_learning_variables.csv'))
df1 = df1[['UNITID', 'INSTNM']]
df1.head()
```

```python
df1.nunique()
```

```python
df2 = pd.read_csv(os.path.join('FourYearUniversities.csv'))
df2 = df2[['UNITID']]
df2.head()
```

```python
df2.nunique()
```

```python
df3 = pd.merge(df2,df1, on='UNITID', how='inner')
```

```python
df3.head()
```

```python
df3.nunique()
```

```python
# df3.to_csv(os.path.join('API_SQL_UNITID-NAME.csv'))
```

```python
metadata1 = pd.read_csv(os.path.join('metadata_needs_join.csv'))
metadata1.head()
```

```python
metadata = pd.merge(metadata1,df3, on='UNITID', how='inner')
```

```python
metadata.head()
```

```python
metadata.shape
```

```python
## Now that the metadata is alltogether, I need to join with the ml_data.csv which contains the machine learning data for prediction. 
# Import ml_data.csv
ml_data = pd.read_csv(os.path.join('ml_data.csv'))
ml_data.head()
```

```python
ml_data.shape
```

```python
final = pd.merge(metadata,ml_data, on='UNITID', how='inner')
```

```python
final_export = final[['UNITID',
                      'INSTNM',
                      'CONTROL',
                      'TUITIONFEE_IN',
                      'TUITIONFEE_OUT',
                      'SAT_AVG_ALL',
                      'TUITFTE',
                      'GRAD_DEBT_MDN',
                      'FAMINC',
                      'MEAN_EARN_6',
                      'MEAN_EARN_8',
                      'MEAN_EARN_10',
                      'ADM_RATE_ALL',
                      'AVGFACSAL',
                      'RET_FT4',
                      'CDR3',
                      'AGE_ENTRY',
                      'UGDS_MEN']]
```
