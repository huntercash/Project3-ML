```python
import pandas as pd
import numpy as np
import glob
import os
import datetime
```

```python
files_path = "Data/CollegeScorecard_Raw_Data"
read_files = glob.glob(os.path.join(files_path,"*.csv"))

college_data = []
for file in read_files:
#     print(file)
    year = file[-14:-7]
    year_data = pd.DataFrame(pd.read_csv(file))
    year_data['YEAR'] = year
    
    year_data = year_data[['UNITID','INSTNM','CITY', 'ZIP','INSTURL','ADM_RATE','ADM_RATE_ALL', 'TUITIONFEE_IN','YEAR']]
    
    college_data.append(year_data)

    
```
